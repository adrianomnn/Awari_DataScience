```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 433 Ã© o cÃ³digo do IPCA

url = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.433/dados?formato=json'
df = pd.read_json(url)
df['data'] = pd.to_datetime(df['data'], dayfirst=True)
df
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
sns.lineplot(y=df['valor'], x=df['data'])
ax.set(title = "SÃ©rie HistÃ³rica IPCA", xlabel = "Ano")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.ylabel('IPCA')
plt.grid(color='gray', linestyle='--', linewidth=0.5, axis='y')
```
