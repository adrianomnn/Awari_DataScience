```python
import pandas as pd
```

```python
#transformou todos os dados no tipo string
df = pd.read_csv('banco_alugueis.csv',dtype = str)
#retirou uma coluna, a primeira, pois ele criou uma coluna amais de índices
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df.head(10)
```

```python
df.columns
```

```python
all_columns = list(df)
df[all_columns] = df[all_columns].astype(str)
```

```python
df.dtypes
```

```python
#está usando uma função regex para pegar só o número, tirando um m2. No entanto, ainda é uma string.
#Olhar o site regexr.com, que explica o funcionamento.
df['area'] = df['area'].str.extract('(.*) m²')
```

```python
df.head()
```

```python
df['vaga'] = df['vaga'].str.extract('(.*) Vaga')
```

```python
df.head()
```

```python
df['banheiro'] = df['banheiro'].str.extract('(.*) Banheiro')
df['quartos'] = df['quartos'].str.extract('(.*) Quarto')
df.head()
```

```python
df.shape
```

```python
df['aluguel'] = df['aluguel'].str.extract('R\$ (.*) /Mês')
#precisa remover esse ponto dos números, porque agora é uma string, e quando voltar a ser número, vai dar confusão (ele vai identificar como decimal)
#regex=True, para dizer à função replace que é um regex
df['aluguel'] = df['aluguel'].replace('\.', '', regex=True)
df.head()
```

```python
df['bairro'].unique()
```

```python
#como tornou tudo em string, mesmo que não tenha nada, espaço vazio, ele conta como uma string, por isso não tem valor valor vazio
df.info()
```
